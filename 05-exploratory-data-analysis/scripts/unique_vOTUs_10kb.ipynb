```python
import pandas as pd
import numpy as np
```

```python
with open('data/stordalen_vOTU_headers_no_carrots.txt') as f:
    refs = [x.strip('\n') for x in f.readlines()]
```

```python
with open('data/all_seqs.fna.self-blastn.clusters_10kb.tsv') as f:
    clusters = f.readlines()
```

```python
count = 0
for i in clusters:
    x = False
    for j in refs:
        if j in i:
            x = True
    if x:
        count += 1
print(count)
```
